## Scipy:

We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:

1. fitting it to the periodic function

2. plot the fit
Data

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
%matplotlib inline
```

```python
max_values=[39,41,43,47,49,51,45,38,37,29,27,25]
min_values=[21,23,27,28,32,31,28,26,21,19,17,18]
```

```python
months = np.arange(12)
```

```python
plt.plot(months,max_values,'ro')
plt.plot(months,min_values,'bo')
plt.xlabel('Months')
plt.ylabel('minmum and maximum temperature')
plt.show()
```

```python
def periodic_function(times,avg,ampl,time_offset):
     return (avg
            + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))
    
```

```python
res_max, cov_max = optimize.curve_fit(periodic_function, months,
                                      max_values, [40, 20, 0])
```

```python
res_min,cov_min=optimize.curve_fit(periodic_function, months,
                                      min_values, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)
```

```python
plt.figure()
plt.plot(months, max_values, 'ro')
plt.plot(days, periodic_function(days, *res_max), 'r-')
plt.plot(months, min_values, 'bo')
plt.plot(days, periodic_function(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```

## 1. Create a pie chart presenting the male/female proportion

```python
import matplotlib.pyplot as plt

genderData = titanic_servey['sex'].value_counts()
print('Gender Data : \n',str(genderData))

fig=plt.figure(figsize=(8,5))
axes=fig.add_axes((0,0,0.5,1))
axes.set_title('Gender Distribution')
labels = 'Male', 'Female'
colors=['blue','pink']
explode=(0.1,0)

plt.pie(genderData,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True, startangle=60)
plt.legend(numpoints=1)
plt.axis('equal')
plt.show()
```

## 2.Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
colors1 = np.where(titanic_servey['sex']=='female','pink','blue')

plt.scatter(titanic_servey['fare'],titanic_servey['age'],c=colors1,alpha=0.8,s=50)


plt.title('Scatterplot with the Fare paid and the Age')

plt.xlabel('Fare')

plt.ylabel('Age')

plt.show()
```
